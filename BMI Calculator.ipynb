```python
# Body Mass Index (BMI)
# BMI is calculated by dividing your weight (in kilograms)
# by the square of your height (in meters)

import time

def type_text(text, delay=0.04):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()

def get_weight_unit_height_unit(): 
    # Dictionary to be iterated on to 
    # print on screen for user to select
    weight_unit_dict = {1:'kilogram(kg)', 2:'Pounds(lb)'}
    height_unit_dict = {1:'feet(ft)',2:'centimeter(cm)',3:'meters(m)'}

    while True:
        print("\nSelect a unit to input your weight")

        # Iterate over the dictionary 
        for key,value in weight_unit_dict.items():
            # Print the values and key
            print(f"{key}. {value}")
        
        get_weight_unit = input("").lower()
        
        if get_weight_unit == 'kilogram' or get_weight_unit == 'kg' or get_weight_unit == "1":
            weight_unit = 1
            break
        elif get_weight_unit == 'pounds' or get_weight_unit == 'lb' or get_weight_unit == "2":
            weight_unit = 2
            break
        
    while True:
        # Input to get the unit height will be calculated in
        print("\nSelect a unit to input your height")
    
        # Iterate over the dictionary
        for key,value in height_unit_dict.items():
            # Print the values and key
            print(f"{key}. {value}")
        get_height_unit = input("").lower()
    
        if get_height_unit == 'feet' or get_height_unit == 'ft' or get_height_unit == "1":
            height_unit = 1
            break
        elif get_height_unit == 'centimeter' or get_height_unit == 'cm' or get_height_unit == "2":
            height_unit = 2
            break
        elif get_height_unit == 'meters' or get_height_unit == 'm' or get_height_unit == "3":
            height_unit = 3
            break
    
    return weight_unit, height_unit

# Get weight and height
# Return if in kg and metres respectively
# else convert them to the above unit
def get_weight_height(weight_unit, height_unit):
    # Iterates till the user gets it right
    while True:
        # Error Handling to ensure user input the right value(int or float)
        try:
            # Return Weight if it is in kilogram
            if weight_unit == 1:
                get_weight = float(input("Input your weight in kg"))
                weight = get_weight
            # Convert Pound(lb) to kilogram(kg) and return it
            elif weight_unit == 2:
                # 1 pounds(lb) is approximately 0.45
                get_weight = float(input("Input your weight in pounds(lb)"))
                weight = get_weight * 0.45
            else:
                weight = 0
        # Exception to show the error
        except ValueError:
            print("It must be a real number (int or float)")
            
        # Error Handling to ensure user input the right value(int or float)
        try:    
            # Convert feet to meters 
            # Return it
            if height_unit == 1:
                # - Feet (ft) and Inches (in) to Meters (m): 1 ft ≈ 0.3048 m, 1 in ≈ 0.0254 m
                get_height = input("\nInput your height in feet (E.g 5 ft 9)")
                splitted_height = get_height.split("ft")
                get_height = [int(height) for height in splitted_height]
                height = (get_height[0] * 0.3048) + (get_height[1] * 0.0254)
                
            # Convert Pound(lb) to kilogram(kg) and return it
            elif height_unit == 2:
                get_height = float(input("\nInput your height in cm"))
                height = get_height * 0.01

            # Return Height if it is in meters
            elif height_unit == 3:
                get_height = float(input("\nInput your height in m"))
                height = get_height
            else:
                height = 0

            return weight, height
        # Exception to show the error
        except ValueError:
            print("It must be a real number (int or float)")

def bmi_calculator(get_weight, get_height):
    result = get_weight / (get_height ** 2)
    result = round(result, 2)
    return result

def main():
    while True:
        type_text("""Body Mass Index (BMI)
BMI is calculated by dividing your weight (in kilograms) by the square of your height (in meters).""")
        
        weight_unit, height_unit = get_weight_unit_height_unit()
        weight, height = get_weight_height(weight_unit, height_unit)
        BMI = bmi_calculator(weight, height)

        print(f"Your Body Mass Index (BMI) is: {BMI}")

        type_text("""\nKeep in mind that BMI is not a perfect measure, as it does not take into account muscle mass or body composition. 
However, it can provide a general indication of whether your weight is in a healthy range.
""")
        play_again = input("Do you want to play again (yes/no)")
        
        if play_again == 'no':
            break

if __name__ == '__main__':
    main()
```
